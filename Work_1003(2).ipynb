```python
def remove_outliers(df_housing):
    df_out = pd.DataFrame()
    for i, subdf in df_housing.groupby('district'):
        m =np.mean(subdf.unit_rate)
        st =np.std(subdf.unit_rate)
        reduced_df = subdf[(subdf.unit_rate>(m-st)) & (subdf.unit_rate<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index = True)
    return df_out

df_housing1 = remove_outliers(df_housing1)
df_housing1.describe()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

import collections, operator
RunResult = collections.namedtuple('RunResult', 'model error pred r2')
result_list = []


model_list = [
             LinearRegression(),
             KNeighborsRegressor(n_neighbors=100),
             XGBRegressor(n_estimators=50, max_depth=6, eta = 0.1),
             DecisionTreeRegressor(max_depth = 4),
             RandomForestRegressor(n_estimators=80, max_depth = 4),
            ]
for model in model_list:
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test, pred) # compute MSE
    r2 = r2_score(y_test, pred) # r2 score (r-square score, max 1, can be -ve)
    rmse = np.sqrt(mse)
    result_list.append(RunResult(model, rmse, pred, r2))

result_list.sort(key=operator.attrgetter('error'))

space=' '
print(f'{space*40}RMSE{space*6}R-squared')
for i, result in enumerate(result_list):
    print(f'{i+1:2}. {type(result.model).__name__:30}    {result.error:6.2f}    {result.r2:.2f}')

best_model = result_list[0].model
lowest_error = result_list[0].error
best_pred = result_list[0].pred
```

```python
predictors = ['unit_rate', 'Zone','saleable_area(ft^2)']
target = 'price'
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
Start = 150
Step = 52

def predict(train, test, predictors, target):
    model = RandomForestClassifier(min_samples_split =10, random_state=1)
    model.fit(train[predictors], train[target])
    preds=model.predict(test[predictors])
    return preds
def backtest (data, predictors, target):
    all_preds =[]
    for i in range(Start, data.shape[0], Step):
        train=df_housing1.iloc[:i]
        test=df_housing.iloc[i:(i+Step)]
        all_preds.append(predict(train, test, predictors, target))
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[Start:][target], preds)
```

```python
preds, accuracy_score = backtest(df_housing1, predictors, target)
```
